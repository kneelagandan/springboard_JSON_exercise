These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
url = "https://www.quandl.com/api/v3/datasets/FSE/AFX_X/data.json?" + \
      "&start_date=2017-01-01&end_date=2017-12-31&api_key=" + API_KEY
req = requests.get(url)
```

```python
json_data = req.json()
json_data["dataset_data"]['column_names']
```

```python
open_dict = {row[0]: row[1] for row in json_data["dataset_data"]["data"] if row[1] != None}
print(open_dict)
```

```python
open_max_date, open_max = max(open_dict.items(), key=lambda k: k[1])
open_min_date, open_min = min(open_dict.items(), key=lambda k: k[1])

print("The highest opening price in 2017"+str(round(open_max, 2))+" date is "+open_max_date)
print("The lowest opening price in 2017 "+str(round(open_min, 2))+" date is "+open_min_date)

```

```python
spread_dict = {row[0]: row[2] - row[3] for row in json_data["dataset_data"]["data"]}
print(spread_dict)
```

```python
spread_max_date, spread_max = max(spread_dict.items(), key=lambda k: k[1])

print("The largest daily price spread in 2017 was "+str(round(spread_max, 2))+" date is "+spread_max_date)

```

```python
close_delta_dict = {}
close_prev =  json_data["dataset_data"]["data"][0][4]
print(close_prev)
```

```python
for row in json_data["dataset_data"]["data"][1:]:
    close = row[4]
    close_delta_dict[row[0]] = abs(close_prev - close)
    close_prev = close

print(close_delta_dict)
```

```python
close_delta_max_date, close_delta_max = max(close_delta_dict.items(), key=lambda k:k[1])

print("The largest change between consecutive day closing prices in 2017 was "+str(round(close_delta_max, 2))+" date is "+close_delta_max_date)

```

```python
volume = [row[6] for row in json_data["dataset_data"]["data"]]
print(volume)
```

```python
vol_avg = sum(volume) / len(volume)


print("The average daily traded volume for 2017 was "+str(round(vol_avg, 2))+" Shares")
```
